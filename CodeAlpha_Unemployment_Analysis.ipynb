```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Sample data create - nuv CSV unte pd.read_csv('unemployment.csv') use cheyyi
data = {
    'Year': [2018, 2019, 2020, 2021, 2022, 2023, 2024],
    'Unemployment_Rate': [6.1, 5.8, 8.9, 7.5, 6.7, 6.2, 5.9],
    'Employed': [485, 492, 450, 468, 480, 490, 498],
    'Unemployed': [32, 30, 44, 38, 35, 32, 31],
    'State': ['AP', 'TS', 'AP', 'TS', 'AP', 'TS', 'AP']
}
df = pd.DataFrame(data)

print("Data preview:")
print(df.head())
print("\nData info:")
print(df.describe())

# 2. Graph 1: Line plot - Year vs Unemployment Rate
plt.figure(figsize=(10,5))
sns.lineplot(x='Year', y='Unemployment_Rate', data=df, marker='o', linewidth=2)
plt.title('Year-wise Unemployment Rate Trend', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Unemployment Rate %')
plt.grid(True)
plt.show()

# 3. Graph 2: Bar plot - Employed vs Unemployed
df_melt = df.melt(id_vars='Year', value_vars=['Employed', 'Unemployed'],
                  var_name='Status', value_name='Count')
plt.figure(figsize=(10,5))
sns.barplot(x='Year', y='Count', hue='Status', data=df_melt)
plt.title('Employed vs Unemployed by Year')
plt.xlabel('Year')
plt.ylabel('Number of People in Lakhs')
plt.legend()
plt.show()

# 4. Graph 3: State-wise average unemployment
plt.figure(figsize=(6,4))
sns.barplot(x='State', y='Unemployment_Rate', data=df, estimator=np.mean, hue='State',palette='Set2',legend=False,errorbar=None)
plt.title('Average Unemployment Rate by State')
plt.xlabel('State')
plt.ylabel('Avg Unemployment Rate %')
plt.show()

# 5. Graph 4: Correlation Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(df[['Unemployment_Rate', 'Employed', 'Unemployed']].corr(),
            annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

# 6. Key insights print
print("\nKey Insights:")
print(f"Highest unemployment: {df['Unemployment_Rate'].max()}% in {df.loc[df['Unemployment_Rate'].idxmax(), 'Year']}")
print(f"Lowest unemployment: {df['Unemployment_Rate'].min()}% in {df.loc[df['Unemployment_Rate'].idxmin(), 'Year']}")
print(f"Average unemployment rate: {df['Unemployment_Rate'].mean():.2f}%")
```
